```python
import pandas as pd
import numpy as np
import pymysql
```

```python
def SQL_load(sql):
    # db접속
    db=pymysql.connect(host='localhost',port=3306,user='root',
                   passwd='661400',db='plt',charset='utf8',
                  cursorclass=pymysql.cursors.DictCursor)
    
    cs=db.cursor()
    
    cs.execute(sql)

    res=cs.fetchall()

    df=pd.DataFrame(res)
    
    db.close()
    return df
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df1=SQL_load('''select invoiceno,Description from plt.retail_on_top10 A join plt.log_df B on A.CustomerID=B.customerid''')
df1
```

```python
Transaction = df1.groupby(['invoiceno', 'Description'])['Description'].count().reset_index(name ='Count')
Transaction
```

```python
basket = Transaction.pivot_table(index='invoiceno', columns='Description', values='Count', aggfunc='sum').fillna(0)
basket.head()
```

```python
def trans(x):
    if x<=0:
        return 0
    if x>=1:
        return 1


basket = basket.applymap(trans)
basket.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
item0=apriori(basket,min_support = 0.03,use_colnames=True)
item0
```
