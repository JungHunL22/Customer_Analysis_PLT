```python
import pandas as pd
import pymysql
```

```python
db=pymysql.connect(host='localhost',port=3306,user='root',
                   passwd='661400',db='plt',charset='utf8',
                  cursorclass=pymysql.cursors.DictCursor)
```

```python
cs=db.cursor()
```

```python
sql='''select * from plt.retail_on_top10;'''
cs.execute(sql)
```

```python
res=cs.fetchall()
```

```python
# db.close()
```

```python
month_sales=pd.DataFrame(res)
month_sales.head()
```

```python
ex=month_sales[month_sales['Country']!='United Kingdom']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
palette = sns.color_palette("pastel")
# sns.palplot(palette)
# 한글 문제
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':  # 맥OS 
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # 윈도우
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system...  sorry~~~')
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x='Month_s',y='sales',hue='Country',data=month_sales)
plt.xticks(rotation=-45)
```
