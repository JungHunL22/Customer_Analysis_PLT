```python
import pandas as pd
import numpy as np
import pymysql
```

```python
def SQL_load(sql):
    # db접속
    db=pymysql.connect(host='localhost',port=3306,user='root',
                   passwd='661400',db='plt',charset='utf8',
                  cursorclass=pymysql.cursors.DictCursor)
    
    cs=db.cursor()
    
    cs.execute(sql)

    res=cs.fetchall()

    df=pd.DataFrame(res)
    
    db.close()
    return df
```
